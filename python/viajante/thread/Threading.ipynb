# Threading
Em Python, o pacote que providencia as funcionalidades de thread é chamado threading, e deve ser importado no começo do seu programa: from threading import Thread.

```python
from threading import Thread
import random
```

Segue um exemplo básico, de uma classe chamada Th que implementa Thread e o método run(). O conteúdo do método run será executado em uma thread separada sempre que o método start, definido na classe Thread e herdado pela classe Th.

```python
 class Th(Thread):

                def __init__ (self, pid):
                      Thread.__init__(self)
                      self.pid = pid

                def run(self):
                      for i in range(150):
                          print(f"Olá Mundo! {self.pid} - {i}")
```

**Executando uma única thread**

```python
a = Th(0)
a.start()
```

**Executando múltiplas threads**

```python
a = []
for i  in range(3):
    a.append(Th(i))
    a[i].start()
```

**Controlando o Fluxo das Trheads**

```python
import random

class Th(Thread):
    def __init__ (self, pid):
        Thread.__init__(self)
        self.pid = pid

    def run(self):
        self.num = random.randint(0, 10)

    def retorno(self):
        return((self.pid, self.num))
```

```python
a = []
for i  in range(30):
    a.append(Th(i))
    a[i].start()
for i  in range(30):
    a[i].join()
    print(f'Thread: {a[i].retorno()[0]} Valor: {a[i].retorno()[1]}')
```

Exercício: Criar um somador de listas em paralelo utilizando as Threads em Python

```python
import random

random.seed(1)
lista = []
for i in range(10000):
    lista.append(random.randint(0, 1000))
print([lista[i] for i in range(10)])
```

```python
from threading import Thread

class Th(Thread):
    lista = []
    total = 0
    def __init__(self, tid, lista):
        Thread.__init__(self)
        self.pid = tid
        self.lista = lista

    def run(self):
        self.total = sum(self.lista)

    def get_total(self):
        return (self.pid, self.total)
```

```python
th = Th(0, lista)
th.start()
total = th.get_total()
print(total)
```

```python
lista0 = []
lista1 = []

for i in range(len(lista) // 2):
    lista0.append(lista[i])
    lista1.append(lista[len(lista)-1 - i])
```

```python
th0 = Th(0, lista0)
th1 = Th(1, lista1)

th0.start()
th1.start()

print(f'A Thread {th0.get_total()[0]} totalizou {th0.get_total()[1]}')
print(f'A Thread {th1.get_total()[0]} totalizou {th1.get_total()[1]}')


total = th0.get_total()[1]+ th1.get_total()[1]
print(f'O resultado final foi: {total}')
```
